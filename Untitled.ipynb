```python
import pandas as pd
import numpy as np
import plotly.express as plt
import plotly.graph_objects as pg
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
sns.set(style = 'darkgrid')
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('salaries_clean.csv')
data.head()
```

```python
plt.figure(figsize = (6,3))
sns.distplot(x = data.total_experience_years)
```

```python
plt.figure(figsize = (8,5))
sns.countplot(x = data.job_title_category)
```

```python
sns.boxplot(x = data.employer_experience_years)
```

```python
#what titile category is note paid
plt.figure(figsize = (12,4))
data = data.loc[data.job_title_category != 'Other']
sns.barplot(x = data.job_title_category, y = data.annual_base_pay, ci = 0, palette = 'Set2', estimator = 'mean')
```

```python
#in canada, us, india, what countries pays data scientist the most
sel = data[['location_country','job_title_category','annual_base_pay']]
sel = sel.loc[(sel.job_title_category == 'Data') & (sel.location_country.isin(['CA','US','IN']))]
sel = sel.groupby('location_country')[['annual_base_pay']].mean().sort_values(by = 'annual_base_pay', ascending = False)

plt.figure(figsize = (6,4))
fig = sns.barplot(x = sel.index, y = sel.annual_base_pay, ci = 0)
fig.bar_label(fig.containers[0], fmt = '%.1f')
```

```python

# Create a dataframe showing only the job title and the salary of all 'Software Developers' in california and Oklahoma.
# data.loc[(data.job_title == 'software developer') & ((data.location_state == 'CA') | (data.location_state == 'OK'))][['job_title', 'annual_base_pay']]
sel = data[['job_title', 'location_state', 'annual_base_pay']]
sel = sel.loc[(sel.job_title == 'software developer') & ((sel.location_state == 'FL') | (sel.location_state == 'CA'))]

fig = sns.barplot(x = sel.location_state, y = sel.annual_base_pay, ci = 0)
fig.bar_label(fig.containers[0], fmt = '%1.f')
plt.title('Software developers in Florida and California')



```
